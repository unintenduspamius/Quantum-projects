```python
import random
import time

def measure_qubit():
    print("Measuring your qubit...")
    time.sleep(1)
    result = random.choice(["|0⟩", "|1⟩"])

    if result == "|0⟩":
        print("""
         _______
        |       |
        |  |0⟩  |
        |_______|
        """)
    else:
        print("""
         _______
        |       |
        |  |1⟩  |
        |_______|
        """)

    print("Qubit collapsed to:", result)

# Run the function
measure_qubit()
```

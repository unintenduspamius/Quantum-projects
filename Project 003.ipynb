```python
def apply_gate(qubit, gate):
    if gate == "X":  # Pauli-X gate (like NOT)
        return "|1⟩" if qubit == "|0⟩" else "|0⟩"
    elif gate == "H":  # Hadamard gate (superposition)
        return "(|0⟩ + |1⟩) / √2"
    elif gate == "I":  # Identity gate
        return qubit
    else:
        return "Invalid Gate"

# Test the function with different gates
initial_qubit = "|0⟩"
print("Initial Qubit:", initial_qubit)

# Try Pauli-X gate
print("After Pauli-X:", apply_gate(initial_qubit, "X"))

# Try Hadamard gate
print("After Hadamard:", apply_gate(initial_qubit, "H"))

# Try Identity gate
print("After Identity:", apply_gate(initial_qubit, "I"))
```
